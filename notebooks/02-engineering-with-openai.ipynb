## Introdução ao prompt engineering

Prompt Engineering: Refere-se a elaboração de instruções a grande modelos de linguagem LLM, para obter as respostas desejadas.
Cada messagem da API tem uma das três funções:

- User: As mensagens do usuário sao avisos do usuário.
- Assistant: As mensagens do assistente são as respostas aos prompts do usuário
- system: As mensagens do sistema fornecem instruções para orientar o comportamento do modelo durante uma conversa.

### Controle de parametros para auxio do modelo:
#### temperature
controla a aleatóriedade das repostas.
- temperature: 0 mais deterministico.
- temperature: 2 altamente aleatório.
#### max_tokens
Controla o comprimento das respostas desejadas.


```python
from dotenv import load_dotenv
import os
from openai import OpenAI

load_dotenv("../.env")

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

conversation_messages = [
    {"role": "system", "content": "Você é um(a) assistente de gestão de eventos muito prestativo(a)."},
    {"role": "user", "content": "Quais são algumas boas maneiras de iniciar uma conversa em eventos para networking?"},
    {"role": "assistant", "content": ""}
]

response = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=conversation_messages
)
print(response.choices[0].message.content)
```

```python
from dotenv import load_dotenv
import os
from openai import OpenAI

load_dotenv("../.env")

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def get_response(prompt):
  response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}], 
    temperature = 0
  )
  return response.choices[0].message.content

response = get_response("Quem descobriu o brasil?")
print(response)
```
